```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load the trained model with custom classes
model = YOLO('yolov8n_trained.pt')

# Load the custom class names
class_names = ['Glass', 'PMD', 'Paper', 'Rest']

# Load an image for inference
img = cv2.imread('testphotos/test5.png')  # Adjust path as necessary

# Perform inference with a low confidence threshold
results = model(img, conf=0.001)  # Set a very low confidence threshold

# Access the first result
result = results[0]

# Check if there are any detections
if result.boxes is not None:
    # Get the bounding boxes, classes, and scores
    boxes = result.boxes.xyxy.cpu().numpy()  # Bounding box coordinates
    classes = result.boxes.cls.cpu().numpy()  # Class IDs
    scores = result.boxes.conf.cpu().numpy()  # Confidence scores

    # Draw bounding boxes on the image
    for box, cls, score in zip(boxes, classes, scores):
        if int(cls) < len(class_names):  # Ensure class ID is within range
            x1, y1, x2, y2 = map(int, box)
            label = f'{class_names[int(cls)]}: {score:.2f}'
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Draw bounding box
            cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)  # Draw label

# Convert BGR to RGB for matplotlib display
rendered_img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the results
plt.imshow(rendered_img_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python
import cv2
from ultralytics import YOLO

# Load the fine-tuned YOLOv8 model
model = YOLO('yolov8n_trained.pt')

# Open the webcam
cap = cv2.VideoCapture(0)  # Try changing to cv2.VideoCapture(1) if you have multiple cameras

# Check if the webcam is opened correctly
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Define the custom class names (adjust according to your dataset)
class_names = ['Glass', 'PMD', 'Paper', 'Rest']

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret:
        print("Error: Could not read frame.")
        break

    # Perform inference with a reasonable confidence threshold
    results = model(frame, conf=0.25)  # Adjust the confidence threshold as needed

    # Access the first result
    result = results[0]

    # Check if there are any detections
    if result.boxes is not None:
        # Get the bounding boxes, classes, and scores
        boxes = result.boxes.xyxy.cpu().numpy()  # Bounding box coordinates
        classes = result.boxes.cls.cpu().numpy()  # Class IDs
        scores = result.boxes.conf.cpu().numpy()  # Confidence scores

        # Draw bounding boxes on the frame
        for box, cls, score in zip(boxes, classes, scores):
            if int(cls) < len(class_names):  # Ensure class ID is within range
                x1, y1, x2, y2 = map(int, box)
                label = f'{class_names[int(cls)]}: {score:.2f}'
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Draw bounding box
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)  # Draw label

    # Display the resulting frame
    cv2.imshow('Live Object Detection', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os

# Load the best trained model
model = YOLO('runs/detect/train8/weights/yolo_finetuned.pt')

# Path to the test images directory
test_images_dir = 'testphotos'

# Function to visualize predictions
def visualize_predictions(predictions, img_path):
    # Read the image
    img = cv2.imread(img_path)
    if img is None:
        print(f"Failed to load image {img_path}")
        return
    
    print(f"Image loaded: {img_path}, shape: {img.shape}")
    
    for pred in predictions:
        if len(pred.boxes) == 0:
            print("No boxes found in the prediction.")
            return
        
        boxes = pred.boxes.xyxy  # Bounding box coordinates
        confs = pred.boxes.conf  # Confidence scores
        cls = pred.boxes.cls  # Class predictions

        for i in range(len(boxes)):
            x1, y1, x2, y2 = map(int, boxes[i])
            label = model.names[int(cls[i])]
            confidence = confs[i]
            # Draw the box
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            # Draw the label and confidence
            cv2.putText(img, f'{label}: {confidence:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    # Convert BGR image to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Display the image
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Loop through all images in the test directory
for img_file in os.listdir(test_images_dir):
    if img_file.endswith(('.png', '.jpg', '.jpeg')):  # Add other extensions if needed
        img_path = os.path.join(test_images_dir, img_file)
        # Get predictions on the image
        predictions = model.predict(source=img_path, save=False)
        # Visualize predictions for the image
        visualize_predictions(predictions, img_path)
```
